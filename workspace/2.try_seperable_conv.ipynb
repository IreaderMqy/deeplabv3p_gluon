```python
def compute_same_padding(kernel_size, dilation, ):
    kernel_size_effective = kernel_size + (kernel_size - 1) * (dilation - 1)
    pad_total = kernel_size_effective - 1
    pad_beg = pad_total // 2
    pad_end = pad_total - pad_beg
    return pad_beg, pad_end
```

```python
class SeparableConv(nn.HybridBlock):
    
    def __init__(self, out_filters, in_filters, kernel_size, strides, dilation, depth_activation):
        super(SeparableConv,self).__init__()
        padding = compute_same_padding(kernel_size, dilation)
        self.depthwise_conv = nn.Conv2D(out_filters, in_channels=in_filters, groups=in_filters, 
                                        dilation=dilation,use_bias=False,
                                        padding=padding, strides=strides,
                                        kernel_size=kernel_size)
        self.bn1 = nn.BatchNorm(axis=1)
        self.pointwise_conv = nn.Conv2D(out_filters, kernel_size=1,use_bias=False)
        self.bn2 = nn.BatchNorm(axis=1)
        self.depth_activation = depth_activation
        
    def hybrid_forward(self, F, x):
        if not self.depth_activation:
            x = F.relu(x)
        x = self.depthwise_conv(x)
        x = self.bn1(x)
        if self.depth_activation:
            x = F.relu(x)
        x = self.pointwise_conv(x)
        x = self.bn2(x)
        if self.depth_activation:
            x = F.relu(x)
        return x     
```
